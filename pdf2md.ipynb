```python
# %%
import os
import subprocess
from tqdm import tqdm

# 股票文件夹的根目录
stock_root = "Eastmoney_report_pdf_download\\reports_pdf"
# 输出根目录
output_root = "pdf2md"
os.makedirs(output_root, exist_ok=True)

# %%
def is_converted(output_path):
    """
    判断一个股票文件夹是否已经转换过：
    只要输出目录下存在至少一个 .md 文件，就认为已转换
    """
    if not os.path.exists(output_path):
        return False
    for root, dirs, files in os.walk(output_path):
        for f in files:
            if f.lower().endswith(".md"):
                return True
    return False

def convert_stock_folder(stock_folder_path, output_path):
    """
    使用 mineru 将一个股票文件夹（含子文件夹）中的所有 PDF 转换为 Markdown
    """
    cmd = [
        "mineru",
        "-p", stock_folder_path,
        "-o", output_path,
        "--backend", "pipeline",
        "--device", "cuda",
        "--batch-size", "8"
    ]
    print("运行命令:", " ".join(cmd))
    subprocess.run(cmd, check=True)

# %%
# 获取所有股票文件夹（reports_pdf 下的一级子目录）
stock_folders = [
    os.path.join(stock_root, name)
    for name in os.listdir(stock_root)
    if os.path.isdir(os.path.join(stock_root, name))
]

# %%
# 逐个处理每只股票文件夹，并显示进度条
for stock_path in tqdm(stock_folders, desc="股票转换进度", unit="stock"):
    stock_name = os.path.basename(stock_path)
    output_path = os.path.join(output_root, stock_name)
    os.makedirs(output_path, exist_ok=True)
 
    # 检查是否已转换过
    if is_converted(output_path):
        print(f"⏩ 跳过 {stock_name}，已存在转换结果")
        continue

    convert_stock_folder(stock_path, output_path)

print("✅ 所有股票报告已转换完成，结果保存在:", output_root)
```

```python
import os
import shutil

# 股票文件夹的根目录
stock_root = "Eastmoney_report_pdf_download/reports_pdf"

def move_deep_reports(stock_root):
    """
    将所有股票文件夹中的 '深度报告' 子文件夹里的 PDF 文件移到外层，
    并删除空的 '深度报告' 文件夹
    """
    # 遍历所有股票文件夹
    for stock_name in os.listdir(stock_root):
        stock_path = os.path.join(stock_root, stock_name)
        if not os.path.isdir(stock_path):
            continue

        deep_report_path = os.path.join(stock_path, "深度报告")
        if os.path.exists(deep_report_path) and os.path.isdir(deep_report_path):
            # 遍历深度报告文件夹中的文件
            for file in os.listdir(deep_report_path):
                file_path = os.path.join(deep_report_path, file)
                if file.lower().endswith(".pdf"):
                    # 移动到股票文件夹顶层
                    target_path = os.path.join(stock_path, file)
                    print(f"移动: {file_path} -> {target_path}")
                    shutil.move(file_path, target_path)

            # 如果深度报告文件夹为空，删除它
            if not os.listdir(deep_report_path):
                print(f"删除空文件夹: {deep_report_path}")
                os.rmdir(deep_report_path)

# 执行
move_deep_reports(stock_root)
print("✅ 所有深度报告文件已移到外层，空文件夹已删除")
```

```python
import os
import re

# 输入 md 文件目录
md_root = "pdf2md"
# 输出 txt 文件目录
txt_root = "md2txt"
# 输出清洗后的 md 文件目录（只保存表格和图像）
md_clean_root = "md_remain"

os.makedirs(txt_root, exist_ok=True)
os.makedirs(md_clean_root, exist_ok=True)
import re

# def clean_line(line):
#     """
#     清理单行文本：
#     1. 修复百分数格式 ($6 3 . 2 \%$ -> 63.2%)
#     2. 删除残留碎片 (如 le_Summary])
#     3. 删除所有 []，保留中文内容，删除英文内容
#     """

#     # 1. 修复百分数：匹配 $...%$
#     line = re.sub( r"\$\s*([-]?\s*[\d\s\.]+)\s*\\%\s*\$", lambda m: m.group(1).replace(" ", "") + "%", line )

#     # 2. 去掉残留碎片
#     if "Summary]" in line or line.strip() in ["]", "["]:
#         return ""  # 删除该行

#     # 3. 删除所有 []，保留中文内容，删除英文内容
#     def bracket_replacer(match):
#         content = match.group(1)
#         # 如果全是英文或符号，就删除
#         if re.fullmatch(r"[A-Za-z0-9\s\-\_]+", content):
#             return ""
#         else:
#             # 保留中文内容，去掉方括号
#             return content

#     line = re.sub(r"\[(.*?)\]", bracket_replacer, line)

#     return line

import re

def clean_line(line):
    """
    清理单行文本：
    1. 修复百分数格式 ($- 5 \%$ -> -5%)
    2. 删除残留碎片 (如 le_Summary])
    3. 删除图像行 (![](images/...))
    4. 删除所有 []，保留中文内容，删除英文内容
    5. 对标题 (# 开头) 去除英文，只保留中文和数字
    """

    # 1. 修复百分数：允许负号和空格
    line = re.sub( r"\$\s*([-]?\s*[\d\s\.]+)\s*\\%\s*\$", lambda m: m.group(1).replace(" ", "") + "%", line )


    # # 2. 去掉残留碎片
    # if "Summary]" in line:
    #     return ""  # 删除该行

    # 3. 去掉图像行
    if re.search(r"!\[.*\]\(.*images.*\)", line):
        return ""  # 删除整行

    # 4. 方括号处理
    def bracket_replacer(match):
        content = match.group(1)
        # 如果全是英文或符号，就删除
        if re.fullmatch(r"[A-Za-z0-9\s\-\_]+", content):
            return ""
        else:
            return content  # 保留中文内容，去掉方括号

    line = re.sub(r"\[(.*?)\]", bracket_replacer, line)

    # # 5. 如果是标题行，去掉英文
    # if line.strip().startswith("#"):
    #     line = re.sub(r"[A-Za-z]+", "", line)

    return line



def split_md_content(md_file, txt_file, md_clean_file):
    """
    将 md 文件拆分为：
    - txt_file: 只保留正文文字
    - md_clean_file: 保存表格和图像内容
    """
    with open(md_file, "r", encoding="utf-8") as f:
        lines = f.readlines()

    text_lines = []
    table_image_lines = []

    for line in lines:
        line = clean_line(line) # 先清理格式
        if not line.strip(): 
            continue
        
        # 判断是否是表格或图像
        if (
            line.strip().startswith("|")
            or line.strip().startswith("<table")
            or line.strip().startswith("</table")
            or re.search(r"!\[.*\]\(.*\)", line)
            or "<img" in line
        ):
            table_image_lines.append(line)
        else:
            text_lines.append(line)

    # 保存 txt 文件（正文）
    with open(txt_file, "w", encoding="utf-8") as f:
        f.writelines(text_lines)

    # 保存新的 md 文件（表格和图像）
    with open(md_clean_file, "w", encoding="utf-8") as f:
        f.writelines(table_image_lines)

# 遍历 md_root 下的所有 md 文件
for root, dirs, files in os.walk(md_root):
    rel_path = os.path.relpath(root, md_root)
    txt_dir = os.path.join(txt_root, rel_path)
    md_clean_dir = os.path.join(md_clean_root, rel_path)

    os.makedirs(txt_dir, exist_ok=True)
    os.makedirs(md_clean_dir, exist_ok=True)

    for file in files:
        if file.lower().endswith(".md"):
            md_file = os.path.join(root, file)
            txt_file = os.path.join(txt_dir, os.path.splitext(file)[0] + ".txt")
            md_clean_file = os.path.join(md_clean_dir, file)

            print(f"处理: {md_file}")
            split_md_content(md_file, txt_file, md_clean_file)

print("✅ 所有 md 文件已拆分完成，txt 中只保留正文，md_clean 中保存表格和图像")
```
